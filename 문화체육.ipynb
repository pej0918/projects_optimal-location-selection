```python
import pandas as pd
import numpy as np
```

```python
df_park=pd.read_excel('서울시 공원 현황.xlsx')
df_park=df_park.iloc[3:,1:3]
df_park.head()
```

```python
df_park.rename(columns={'자치구':'자치구','합계':'공원 개수'},inplace=True)
df_park.head()
```

```python
df_phy=pd.read_excel('서울시 공공체육시설 현황.xlsx')
df_phy.head()
```

```python
df_phy=df_phy.iloc[4:,1:3]
df_phy.rename({'자치구':'자치구', '공공체육시설':'공공체육시설 개수'})
df_phy.head()
```

```python
df_lib=pd.read_excel('서울시 도서관 현황.xlsx')
df_lib.head()
```

```python
df_lib.replace('-', 0,inplace=True)
```

```python
df_lib.info()
```

```python
for i in range(len(df_lib['자치구'])):
    df_lib.loc[i, '도서관 개수']=df_lib.loc[i,'국립도서관']+df_lib.loc[i,'공공도서관']
```

```python
df_lib=df_lib.loc[1:,['자치구','도서관 개수']]
df_lib.head()
```

```python
df_welf=pd.read_excel('서울시 지역문화복지시설 현황.xlsx')
df_welf.head()
```

```python
df_welf=df_welf.replace('-',0)
df_park=df_park.replace('-',0)
df_phy=df_phy.replace('-',0)
```

```python
df_welf=df_welf.iloc[:,1:]
df_welf['복지시설']=df_welf.sum(axis=1)
df_welf=df_welf.loc[:,['자치구','복지시설']]
df_welf.head()
```

```python
df_lib.info()
```

```python
df_phy.info()
```

```python
df_phy['공공체육시설']=df_phy['공공체육시설'].astype(int)
df_phy.info()
```

```python
df_phy=df_phy.iloc[:-1,:]
df_phy.info()
```

```python
df_park.info()
```

```python
df_park
```

```python
df_park=df_park.iloc[:-1,:]
df_park['공원 개수']=df_park['공원 개수'].astype(int)
df_park.info()
```

```python
df=pd.merge(df_park,df_lib, how='outer',on='자치구')
```

```python
df=pd.merge(df,df_phy, how='outer',on='자치구')
```

```python
df=pd.merge(df,df_welf, how='outer',on='자치구')
```

```python
df.head()
```

```python
df['문화/체육시설 합계']=df.sum(axis=1)
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

for i in ['공원 개수','도서관 개수','공공체육시설','복지시설','문화/체육시설 합계']:
    scaler=MinMaxScaler(feature_range=(1,10))
    df[i]=scaler.fit_transform(df[i].values.reshape(-1,1))
```

```python
df.head()
```

```python
df['문화/체육 지수']=0.3*df['공원 개수'] + 0.5 *df['복지시설'] + 0.1*df['공공체육시설'] + 0.1*df['도서관 개수']
```

```python
df.head()
```

```python
df=df.loc[:,['자치구','문화/체육 지수']]
df.head()
```

```python
df.to_csv('서울시_문화_체육 지수.csv',index=False)
```
