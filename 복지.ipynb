```python
import pandas as pd
import numpy as np
```

```python
df_wel=pd.read_csv('서울시 사회복지시설_자치구.csv',encoding='cp949')
df_wel.head()
```

```python
df_wel.shape
```

```python
df_wel=df_wel.groupby('시군구명').count()
df_wel.head()
```

```python
df_police=pd.read_excel('서울시 치안시설 현황.xlsx')
df_police.head()
```

```python
df_police=df_police.replace('-',0)
df_police=df_police.iloc[:,1:]
df_police.head()
```

```python
df_police['치안시설(경찰/소방)']=df_police.sum(axis=1)
df_police.head()
```

```python
df_police=df_police.loc[:,['자치구','치안시설(경찰/소방)']]
df_police.head()
```

```python
df_wel=df_wel.reset_index()
df_wel=df_wel.rename(columns={'시군구명':'자치구','시설명':'사회복지시설'})
df_wel.head()
```

```python
df=pd.merge(df_wel,df_police,how='outer',on='자치구')
```

```python
df
```

```python
df.drop(14,inplace=True)
```

```python
df=df.reset_index()
```

```python
df=df.iloc[:,1:]
df.head()
```

```python
df.info()
```

```python
from sklearn.preprocessing import MinMaxScaler

for i in ['사회복지시설','치안시설(경찰/소방)']:
    scaler=MinMaxScaler(feature_range=(1,10))
    df[i]=scaler.fit_transform(df[i].values.reshape(-1,1))
df.head()
```

```python
df['복지 지수']=0.5 *df['사회복지시설'] + 0.5*df['치안시설(경찰/소방)']
df.head()
```

```python
df=df.loc[:,['자치구','복지 지수']]
df.head()
```

```python
df.to_csv('서울시_복지지수.csv',index=False)
```
